```python
def count_num_distri(substanceName, dataframe, col):
    data = dataframe.loc[dataframe['SubstanceName'] == substanceName]
    data_dict = {}
    null_dict = {}
    
    for location in list(data[col]):
        if sum(data.loc[data[col] == location]['DrugReports']) != 0:
            data_dict[location] = sum(data.loc[data[col] == location]['DrugReports'])
        else:
            null_dict[substanceName] = [location]
    data_dict = sorted(data_dict.items(), key=operator.itemgetter(1), reverse=True)
    return data_dict, null_dict
```

```python
distri_dict = {}
null_substance_dict = {}
for substance in substance_lst:
    count_dict, null_dict = count_num_distri(substance, nflis_data, 'State')
    distri_dict[substance] = count_dict
    null_substance_dict.update(null_dict)
distri_dict
```

## Substance Distribution by States

```python
state_distri_by_year_dict = {}
state_null_substance_dict = {}
for year in sorted(year_lst):
    year_data = nflis_data.loc[nflis_data['Year'] == year]
    distri_dict = {}
    for substance in substance_lst:
        count_dict, null_dict = count_num_distri(substance, year_data, 'State')
        distri_dict[substance] = count_dict
        state_null_substance_dict.update(null_dict)
    state_distri_by_year_dict[year] = distri_dict
state_distri_by_year_dict
```

## Substance Distribution by County

```python
def count_num_distri_county(substanceName, dataframe):
    data = dataframe.loc[dataframe['SubstanceName'] == substanceName]
    data_dict = {}
    for state in list(data['State']):
        state_data = data.loc[data['State'] == state]
        state_dict = {}
        for county in list(state_data['County']):
            state_dict[county] = sum(state_data.loc[state_data['County'] == county]['DrugReports'])
        data_dict[state] = sorted(state_dict.items(), key=operator.itemgetter(1), reverse=True)
    return data_dict
```

```python
county_distri_by_year_dict = {}
for year in sorted(year_lst):
    year_data = nflis_data.loc[nflis_data['Year'] == year]
    distri_dict = {}
    for substance in substance_lst:
        distri_dict[substance] = (count_num_distri_county(substance, year_data))
    county_distri_by_year_dict[year] = distri_dict
county_distri_by_year_dict
```

### Used Substance Every year

```python
used_substance_dict = {}

for year in sorted(year_lst):
    used_lst = []
    year_data = nflis_data.loc[nflis_data['Year'] == year]
    for substance in substance_lst:
        substance_data = year_data.loc[year_data['SubstanceName'] == substance]
        if sum(substance_data['DrugReports']) != 0:
            used_lst.append(substance)
    used_substance_dict[year] = used_lst
used_substance_dict
```

```python
used_count_dict = {}
for year in sorted(used_substance_dict):
    used_count_dict[year] = len(used_substance_dict[year])
used_count_dict
```

```python
diff_dict = {}
for curr_year in [2010,2011,2012,2013,2014,2015,2016,2017]:
    lst = [2010,2011,2012,2013,2014,2015,2016,2017]
    lst.remove(curr_year)
    year_dict = {}
    for year in lst:
        year_dict[year] = set(used_substance_dict[year]) -set(used_substance_dict[curr_year])
    diff_dict[curr_year] = year_dict
diff_dict
```

```python
diff_count_dict = {}
for year in sorted(diff_dict):
    temp_dict = diff_dict[year]
    count_dict = {}
    for test_year in temp_dict:
        count_dict[test_year] = len(temp_dict[test_year])
    diff_count_dict[year] = count_dict
diff_count_dict
```

```python
def draw_larger_than_100_plot(dataframe, state): 
    import matplotlib.pyplot as plt
    
    count_dict = {}
    for substance in substance_lst:
        sub_data = dataframe.loc[dataframe['SubstanceName'] == substance]
        if sum(sub_data['DrugReports']) > 100:
            sub_lst = []
            for year in sorted(set(nflis_data['Year'])):
                year_data = sub_data.loc[sub_data['Year'] == year]
                sub_lst.append(sum(year_data['DrugReports']))
            count_dict[substance] = sub_lst

    count_df = pd.DataFrame.from_dict(count_dict)

    plt.style.use('seaborn-darkgrid')
    
    for substance in count_dict:
        plt.plot(substance, data = count_df, label = substance)
    
    plt.legend(bbox_to_anchor=(1,1),loc='upper left', ncol =1)
    plt.xlabel('Year')
    plt.ylabel('Number of Drug Report')
    plt.title('State '+ str(state) + ' Drug Report Trend (with sum larger than 100)')
    plt.xticks([i for i in range(8)], sorted(set(nflis_data['Year'])))

    plt.savefig(str(state) + '_larger_than_100.png', bbox_inches='tight')
    print(str(state) + 'larger than 100 DONEEEE')
```
