```python
nflis_data = pd.read_csv("2018_MCMProblemC_DATA/updated_nflis_data.csv")
nflis_data = nflis_data.assign(location = lambda x: x.County + ", " + x.State)
```

```python
substance_lst =set(nflis_data['SubstanceName'])
```

```python
print(substance_lst)
```

```python
def count_num_distri(substanceName):
    data = nflis_data.loc[nflis_data['SubstanceName'] == substanceName]
    data_dict = {}
    for state in list(data['State']):
        data_dict[state] = sum(data.loc[data['State'] == state]['DrugReports'])
    return data_dict
```

```python
distri_dict = {}
for substance in substance_lst:
    distri_dict[substance] = (count_num_distri(substance))
distri_dict
```
